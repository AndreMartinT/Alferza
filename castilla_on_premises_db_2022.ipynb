### Prototipo de estructura para una base de datos en SQL:



PORQUE UNA BASE DE DATOS SQL 

- Obtener información sobre procesos y mantener un proceso de mejora contínua
- Optimizar procesos para ser más productivos en flujos de trabajo
- Gestionar diferentes líneas de negocio de forma ordenada y estructurada, más eficientes
- SQL permite combinar de forma eficiente diferentes tablas para extraer información relacionada
- Gestiona su seguridad, e integridad de los datos, actualiza sus contenidos, recupera los datos y los comparte entre diferentes usuarios en tiempo real
